### Bucle While

image.png

````python
## Vamos fazer alguns exemplos
## Vamos fazer o exerccio para nmero, mas dentro de uma lista de nmeros
## Vamos parar o ciclo dos loopings, assim que encontrarmos o par

is_even = False
index = 0
numbers = [11, 5, 13, 8, 34]

while not is_even:
    if numbers[index] % 2 == 0:
        is_even = True
        print("Even number found: " + str(numbers[index]))
    
    index += 1
````

````python
## Lembre-se que a condio pode ser qualquer expresso
## Que avalia uma condio lgica, voc pode usar os operadores relacionais ou os lgicos


## O que voc acha que acontece neste caso?
## Ele ser executado para sempre, ento, aps um momento, pare de executar manualmente
while "OPS":
    print("We are inside the loop")
````
