```python
## Aqui estamos criando a Classe Car

class Car:
    def __init__(self, model, brand, color):
        self.model = model
        self.brand = brand
        self.color = color
    
    def run(self):
        print("The car " + self.model + " is running.")
        
toyota = Car('Auris', 'Toyota', 'brown')
toyota.run()
```

**Exerccio**

Os alunos devem incrementar o cdigo com variveis e mtodos, se acharem necessrio e, ao final, imprimir na tela todo o histrico do aluno, com todas as notas e mdias, inclusive informando se o status do aluno foi: "aprovado por mdia", "aprovado na final" ou "reprovado". Consideram que a mdia aritimtica para aprovao  7,0 (sete) e aps final 5,0 (cinco).

```python
class DesempenhoAcademico():

  def __init__(self, aluno):
    self.nomeAluno = aluno
    self.media = 0
    self.situacao = False

  def calcularMedia(self, nota1, nota2):
    self.media = (nota1 + nota2) / 2
    if (self.media >= 7.00):
      self.situacao = True
    return self.media

  def calcularMediaAposFinal(self, notaFinal):
    self.media = (self.media + notaFinal) / 2
    if (self.media >= 5.00):
      self.situacao = True
    return self.media

  def situacaoAluno(self):
    if (self.situacao == True and self.media >= 5.00):
      print("Aluno ", self.nomeAluno, " est em situao Aprovado")
    else:
      print("Aluno ", self.nomeAluno, " est em situao Reprovado")


desempenho = DesempenhoAcademico('Natasha Pelizari')
desempenho.calcularMedia(7, 8)
desempenho.situacaoAluno()
```
